※ 구글 드라이브에서 RGB, HSV, LAB 데이터를 각각 다운로드 후 압축 해제

```python
!pip install gdown
!pip install scipy

import gdown
import zipfile
import os

# 다운로드할 파일들의 URL 배열
urls = [
    '',
    '',
    '',
]

# 각 파일의 파일명 배열
outputs = [
    'RGB.zip',
    'HSV.zip',
    'LAB.zip',
    # 각 URL에 해당하는 파일명으로 변경
]

# 각 파일의 압축 해제 디렉토리 배열
extract_paths = [
    'RGB', # 파일1의 압축을 해제할 디렉토리 경로
    'HSV', # 파일2의 압축을 해제할 디렉토리 경로
    'LAB', # 파일3의 압축을 해제할 디렉토리 경로
    # 필요한 만큼 디렉토리 경로 추가
]

# URL, 파일명, 압축 해제 디렉토리의 개수가 모두 일치하는지 확인
if len(urls) != len(outputs) or len(urls) != len(extract_paths):
    print("URL, 파일명, 압축 해제 디렉토리의 개수가 일치하지 않습니다.")
else:
    # 각 파일에 대해 순회하여 다운로드 및 압축 해제 수행
    for url, output, extract_path in zip(urls, outputs, extract_paths):
        # 파일 다운로드
        print(f"Downloading file from {url}...")
        gdown.download(url, output, quiet=False)
        print(f"File downloaded as {output}")

        # 압축 해제 디렉토리 생성
        if not os.path.exists(extract_path):
            os.makedirs(extract_path)

        # ZIP 파일 압축 해제
        print(f"Extracting {output} to {extract_path}...")
        with zipfile.ZipFile(output, 'r') as zip_ref:
            zip_ref.extractall(extract_path)
        print(f"Extraction completed for {output}")

print("모든 작업 완료")
#커널 재실행!
```

※ 카테고리별 데이터 수의 균형화

```python
#데이터 균형화
import os
import random
from pathlib import Path

def delete_random_files_until_count(folder_path, target_count=5115):
    for root, dirs, files in os.walk(folder_path):
        jpg_files = [f for f in files if f.endswith('.jpg')]
        current_count = len(jpg_files)

        while current_count > target_count:
            file_to_delete = random.choice(jpg_files)
            file_path = Path(root) / file_to_delete
            os.remove(file_path)
            jpg_files.remove(file_to_delete)
            current_count -= 1

def count_jpg_files(folder_path):
    for root, dirs, files in os.walk(folder_path):
        jpg_files = [file for file in files if file.endswith('.jpg')]
        print(f"Folder {root} has {len(jpg_files)} jpg files.")

# 여러 상위 폴더 경로 리스트 설정
top_folders = ['RGB', 'HSV', 'LAB']

for folder_path in top_folders:
    print(f"Processing folder: {folder_path}")
    delete_random_files_until_count(folder_path)
    print(f"Final count of jpg files in {folder_path}:")
    count_jpg_files(folder_path)
```

※ 훈련데이터와 검증데이터를 8:2로 스플릿

```python
import os
import shutil
import random

# 원본 데이터 디렉토리 및 대상 디렉토리 설정
source_dirs = ['RGB', 'HSV', 'LAB']  # 여러 소스 디렉토리
target_dirs = ['RGB_V', 'HSV_V', 'LAB_V']  # 대상 디렉토리

# 비율 설정
split_ratio = 0.2

# 각 폴더별로 작업 수행
for source_dir, target_dir in zip(source_dirs, target_dirs):
    # 각 폴더별로 작업 수행
    for root, dirs, files in os.walk(source_dir):
        for category in dirs:
            category_dir = os.path.join(root, category)

            # 파일 리스트 가져오기
            files = [f for f in os.listdir(category_dir) if os.path.isfile(os.path.join(category_dir, f))]

            # 랜덤하게 파일 선택
            num_files_to_move = int(len(files) * split_ratio)
            files_to_move = random.sample(files, num_files_to_move)

            # 대상 폴더 생성 (카테고리별)
            target_category_dir = os.path.join(target_dir, category)
            os.makedirs(target_category_dir, exist_ok=True)

            # 파일 이동
            for f in files_to_move:
                source_file = os.path.join(category_dir, f)
                target_file = os.path.join(target_category_dir, f)
                shutil.move(source_file, target_file)



print("스플릿 완료")
```

※ CNN 모델 학습

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, Activation
from tensorflow.keras.models import Sequential
from tensorflow.keras.regularizers import l1, l2, l1_l2

#tf.keras.backend.clear_session()

# 기본 경로 설정
train_dir = 'LAB'  # train data directory
validation_dir = 'LAB_V'  # validation data directory

# 훈련 데이터셋을 생성하기 위한 ImageDataGenerator 설정
# train_datagen = ImageDataGenerator(rescale=1./255)

#증강
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,  # -20도에서 20도 사이에서 무작위로 회전
    width_shift_range=0.2,  # 가로 크기의 20% 내에서 무작위로 이동
    height_shift_range=0.2,  # 세로 크기의 20% 내에서 무작위로 이동
    #brightness_range=[0.8, 1.2],  # 밝기를 80%에서 120% 사이로 조절
    horizontal_flip=True,  # 수평으로 뒤집기
    zoom_range=0.2,  # 80%에서 120% 사이로 무작위로 줌
    fill_mode='nearest'  # 회전 또는 너비/높이 이동으로 인해 생성된 공간을 가까운 픽셀 값으로 채움
)

# 검증 데이터셋을 생성하기 위한 ImageDataGenerator 설정
validation_datagen = ImageDataGenerator(rescale=1./255)

# 훈련 데이터셋
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical')

# 검증 데이터셋
validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=False)


# RGB Best Model
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(256, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(1024, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(12, activation='softmax')
])


# 모델 컴파일
model.compile(loss='categorical_crossentropy',
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.00001),
              metrics=['accuracy'])

# 모델 요약 출력
model.summary()

# 콜백 설정
checkpoint_cb = tf.keras.callbacks.ModelCheckpoint("RGB_best.h5", save_best_only=True)

# 모델 학습
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples//train_generator.batch_size,
    epochs=20,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples//validation_generator.batch_size,
    callbacks=[checkpoint_cb])

# 학습된 모델 저장
model.save("RGB.h5")
```

※ 모델 추가 학습

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint

from tensorflow.keras.optimizers import Adam

# 이미 학습된 모델 불러오기
model = load_model('RGB.h5')

# 불러온 모델의 현재 옵티마이저 구성 확인
print(model.optimizer)

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    zoom_range=0.2,
    fill_mode='nearest'
)

validation_datagen = ImageDataGenerator(rescale=1./255)

train_dir = 'RGB'  # 훈련 데이터셋의 경로
validation_dir = 'RGB_V'  # 검증 데이터셋의 경로

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)

print(train_generator.class_indices)

# 새 학습률 설정
new_learning_rate = 0.000001
model.compile(optimizer=Adam(learning_rate=new_learning_rate),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# 추가 학습을 위한 설정
additional_epochs = 80  # 추가로 학습할 에포크 수
checkpoint_cb = ModelCheckpoint("RGB_Best.h5", save_best_only=True)  # 업데이트된 모델 저장

# 추가 학습 시작
model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=additional_epochs,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // validation_generator.batch_size,
    callbacks=[checkpoint_cb]
)

# 추가 학습 후 모델 저장
model.save("RGB.h5")
```

※ 모델 성능 평가

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# 모델 불러오기
model = load_model('RGB.h5')

# 테스트 데이터셋 준비
# 주의: 이 예시는 검증 데이터셋을 테스트로 사용합니다. 실제로는 별도의 테스트 데이터셋을 사용해야 합니다.
test_datagen = ImageDataGenerator(rescale=1./255)
test_dir = 'RGB_V'

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),  # 모델에 맞는 이미지 크기
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)

# 모델 평가
results = model.evaluate(test_generator)
print(f'Test Loss: {results[0]}')
print(f'Test Accuracy: {results[1]}')
```
